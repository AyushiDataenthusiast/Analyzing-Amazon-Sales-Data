```python
#amazon sales data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
dataset['Unit Price']= dataset['Unit Price'].astype('int') # change datatype from float to int
dataset['Unit Cost']= dataset['Unit Cost'].astype('int')
dataset['Total Revenue']= dataset['Total Revenue'].astype('int')
dataset['Total Cost']= dataset['Total Cost'].astype('int')
dataset['Total Profit']= dataset['Total Profit'].astype('int')
dataset['Order Date']= dataset['Order Date'].astype('datetime64[ns]') # change datattype from obj to datetime
dataset['Ship Date']= dataset['Ship Date'].astype('datetime64[ns]')
```

```python
dataset['Unit Price'].dtypes  # to check datatype
dataset['Unit Cost'].dtypes
dataset['Total Revenue'].dtypes
dataset['Total Cost'].dtypes
dataset['Total Profit'].dtypes
dataset['Order Date'].dtypes
dataset['Ship Date'].dtypes
```

```python
dataset['Sales Channel'].unique() # to check the data inside column is unique
```

```python
dataset['Order Year']= dataset['Order Date'].dt.year # column for year
```

```python
dataset['Order Month']= dataset['Order Date'].dt.month_name().str[:3] # column for month
```

```python
dataset.to_csv("Amazon_sales_analysis_transformm.csv", index=False) #  save csv files
```

# EXPLORATORY DATA ANALYSIS

# Total revenue in each year

```python
plt.figure(figsize=(5,4))
sbn.lineplot(x="Order Year",data=dataset,y="Total Revenue",estimator='sum',errorbar=("ci",0))
plt.title("Total Revenue in Year")
plt.plot()
plt.show()
```

Maximum revenue was in year 2012

# Total revenue in Month

```python
plt.figure(figsize=(5,5))
sbn.barplot(x="Order Month",data=dataset,y="Total Revenue",estimator='sum',errorbar=("ci",0))
plt.title("Total Revenue in Month")
plt.plot()
plt.show()
```

Total revenue was highest in the month of February.

# Items highest revenue

```python
plt.figure(figsize=(15,5))
sbn.barplot(x="Item Type",data=dataset,y="Total Revenue",estimator='sum',errorbar=("ci",0))
plt.title("Highest revenue in Items")
plt.plot()
plt.show()
```

We can see Cosmetics items is having Highest revenue following Office supplies and Household,lowest item revenue is for fruits.

# Highest Sales Channel

```python
dataset['Sales Channel'].value_counts() #total counts of sales channnel
```

```python
plt.figure(figsize=(3,4))
sbn.countplot(x="Sales Channel", data=dataset)
plt.title("Count of Sales Channel")
plt.show()
```

Sales are done 50% Online and 50% Offline

# Highest Revenue by region

```python
plt.figure(figsize=(23,8))
sbn.barplot(x="Region",data=dataset,y="Total Revenue",errorbar=("ci",0))
plt.title("Total Revenue by Region")
plt.plot()
plt.show()
```

Highest revenue by region is ASIA following by North America and lowest revenue is in Sub Saharan Africa.
